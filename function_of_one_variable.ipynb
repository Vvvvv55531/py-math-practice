```python
# Импорт библиотек numpy и matplotlib:
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

# Создание интервала:
x_min = 0.01
x_max = 10 # интервал [0.01, 10]
x = np.linspace(x_min, x_max, 1000) # 1000 точек

# Функции y1(x) и y2(x):
y1 = (x**3 + 3 * x**2) / 4 - 5
y2 = 2 / np.sin(0.1 * x)

# Настройка оформления:
plt.figure()
matplotlib.rc("font", size=11)  # размер шрифта надписей

# Построение графиков функций:
plt.plot(x, y1, label="y1(x)")
plt.plot(x, y2, label="y2(x)")

# Подписи осей и заголовок:
plt.xlabel("x")
plt.ylabel("y(x)")
plt.title("Графики функций y1(x) и y2(x) на отрезке [0.01, 10]", \
          fontsize=10) # размер шрифта названия графика

plt.legend() # легенда с подписями графиков

plt.xlim(x_min, x_max)
plt.grid() # сетка
plt.show() # отображение графика
```

```python
# Импорт библиотек numpy и matplotlib:
import numpy as np 
import matplotlib.pyplot as plt 

def graphs(points, x_scale):
    # Создание интервала:
    x = np.linspace(0.01, 10, points) 

    # Функции y1(x) и y2(x):
    y1 = (x**3 + 3 * x**2) / 4 - 5
    y2 = 2 / np.sin(0.1 * x)

    # Функции y3(x) и y4(x):
    y3 = np.sin(y1)
    y4 = np.sin(y2)

    plt.figure(figsize=(10,2)) # размер окна 10 на 2 дюйм 
    
   # Построение графиков функций:
    plt.plot(x, y3, label="y3(x)")
    plt.plot(x, y4, label="y4(x)")

    # Подписи осей:
    plt.xlabel("x") 
    plt.ylabel("y3, y4") 

    # Масштаб оси x:
    if x_scale == "log":
        x_scale = "лог." 
        plt.xscale("log") 
    else: 
        x_scale = "лин." 

    # Заголовок графика:
    graph_title =  ("Графики функций y3(x) и y4(x) " 
        + str(points) + " точек, " + x_scale + " масштаб")
    
    plt.title(graph_title) 
    plt.legend(loc="best") 

# Вызовы функции:
graphs(600, "lin") # слишком мало точек
graphs(20000, "lin") # достаточно для линейного
graphs(20000, "log") # маловато для логарифмического
graphs(500000, "log") # очень подробно, дольше отрисовывается

# Отображение всех графиков сразу
plt.show()
```

```python
# Импорт библиотек numpy и matplotlib:
import numpy as np 
import matplotlib.pyplot as plt 

# Реализация исходных функций:
def y1(x):
    return (x**3 + 3 * x**2) / 4 - 5

def y2(x):
    return 2 / np.sin(0.1 * x)

# Множество значений x от 0 до 10:
x_all = np.linspace(0.01, 10, 10**7) 

# Нахождение M = max|y1(x)|:
y1_v = y1(x_all)
M = np.max(np.abs(y1_v))

# Приближенное нахождение L при M = |y2(L)|:
y2_v = np.abs(y2(x_all))
diff = np.abs(y2_v - M)
idx_L = np.argmin(diff)
L = x_all[idx_L]

# Промежуток x от L до 10
x_range = np.linspace(L, 10, 1000)
y1_range = y1(x_range)
y2_range = y2(x_range)

# Построение графиков:
def graphs(x, y1, y2, M, x_scale):
    plt.figure(figsize=(10, 4))

    # Построение графиков функций:
    plt.plot(x, y1, label="y1(x)", color="b")
    plt.plot(x, y2, label="y2(x)", color="g")
    
    # Масштаб оси x:
    if x_scale == "log":
        plt.xscale("log")
        x_scale = "логарифмический масштаб"
    else:
        x_scale = "линейный масштаб"

    # Подписи осей и заголовок:
    plt.xlabel("x")
    plt.ylabel("y(x)")
    plt.title(f"Графики y1(x) и y2(x) на интервале [{L:.4f}, 10], {x_scale}")
    
    plt.ylim(-M, M) # промежуток от M до -M
    plt.grid(which="major") # сетка
    plt.grid(which="minor", linestyle=":") # доп. сетка
    plt.legend() # легенда

# Линейный масштаб:
graphs(x_range, y1_range, y2_range, M, x_scale="lin")

# Логарифмический масштаб:
graphs(x_range, y1_range, y2_range, M, x_scale="log")

plt.show()

# Вывод значений:
print(f"M = {M:.2f}, L ≈ {L:.4f}")
```
